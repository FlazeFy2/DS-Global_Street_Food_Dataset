```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("global_street_food.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Vegetarian Street Food Comparison
# Dataset
top_context = df['Vegetarian'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Vegetarian Comparison")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("global_street_food.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Price Category Comparison
# Dataset
# Categorize Price into Price Category 
df['Price Category'] = pd.cut(
    df['Typical Price (USD)'], 
    bins=[0, 2.5, 4.5, 6], 
    labels=['Cheap', 'Normal', 'Expensive']
)

top_context = df['Price Category'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Price Category Comparison")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("global_street_food.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Street Food's Continent Comparison
# Dataset - Mapping Country to Continent
mapping = {
    # Asia
    'Japan': 'Asia',
    'India': 'Asia',
    'Thailand': 'Asia',
    'Israel': 'Asia',
    'Lebanon': 'Asia',
    # North America
    'USA': 'North America',
    'Mexico': 'North America',
    # Africa
    'South Africa': 'Africa',
    'Nigeria': 'Africa',
    'Egypt': 'Africa',
    # Europe
    'Turkey': 'Europe'
}

# Create new column "Continent"
df['Continent'] = df['Country'].map(mapping).fillna('Other')

# Dataset
top_context = df['Continent'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Street Food's Continent Comparison")
plt.ylabel("")
plt.show()
```
