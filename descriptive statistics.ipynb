```python
import pandas as pd
import statistics
import numpy as np

# Declaration
# Descriptive Statistic for Analyzed Price for Each Food's Origin Country
data = pd.read_csv("global_street_food.csv")

country = data['Country'].unique().tolist()

results = []

for dt in country:
    country_data = data[data['Country'] == dt]

    total_food = country_data['Typical Price (USD)'].count()
    min_amount = country_data['Typical Price (USD)'].min()
    max_amount = country_data['Typical Price (USD)'].max()
    avg_amount = country_data['Typical Price (USD)'].mean()
    
    results.append({
        'Country': dt,
        'Total Food': total_food,
        'Min Typical Price': min_amount,
        'Max Typical Price': max_amount,
        'Average Typical Price': round(avg_amount,2),
    })

res_df = pd.DataFrame(results)
res_df = res_df.sort_values(by='Average Typical Price', ascending=False)
res_df
```

```python
import pandas as pd
from collections import Counter

# Descriptive Statistic - Analyzing Most Used Ingredient per Country
data = pd.read_csv("global_street_food.csv")

country_results = []

# Get all unique countries
for country in data['Country'].unique():
    country_data = data[data['Country'] == country] 
    all_ingredients = [] 
    
    # Find and Split Ingredient Call Each Country
    for ingredients in country_data['Ingredients'].dropna(): 
        ingredient_list = [i.strip().lower() for i in ingredients.split(',')]  
        all_ingredients.extend(ingredient_list)  
    
    # Count the frequency
    ingredient_counts = Counter(all_ingredients)
    
    # Get the most used ingredient
    if ingredient_counts:
        most_common_ingredient, count = ingredient_counts.most_common(1)[0]  
    else:
        most_common_ingredient, count = None, 0 
    
    country_results.append({
        'Country': country,
        'Most Common Ingredient': most_common_ingredient,
        'Total Used': count
    })

ingredient_per_country_df = pd.DataFrame(country_results)
print(ingredient_per_country_df)
```

```python
import pandas as pd

# Display settings
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.width', 0)

# Descriptive Statistic - Group by Country, Find Average Price and Most Used Cooking Method
data = pd.read_csv("global_street_food.csv")

# Get all unique countries
country = data['Country'].unique().tolist()

results = []
for dt in country:
    # Filter data for this country only
    country_data = data[data['Country'] == dt]
    result = {'Country': dt}

    # Get Typical Price (USD) average for the country
    result['Average Typical Price (USD)'] = country_data['Typical Price (USD)'].mean()

    # Get Most Used Cooking Method for the country
    result['Most Used Cooking Method'] = country_data['Cooking Method'].mode().iloc[0]
    results.append(result)

res_df = pd.DataFrame(results)
print(res_df)
```

```python
import pandas as pd

# Display settings
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.width', 0)

# Descriptive Statistic - Most Expensive and Cheapest Food Per Country
data = pd.read_csv("global_street_food.csv")

# Get all unique countries
country = data['Country'].unique().tolist()

results = []
for dt in country:
    # Filter data for this country only
    country_data = data[data['Country'] == dt]
    result = {'Country': dt}

    # Get Most Expensive Food for the country
    max_idx = country_data['Typical Price (USD)'].idxmax()
    result['Most Expensive Food'] = country_data.loc[max_idx, 'Dish Name']
    result['Most Expensive Price (USD)'] = country_data.loc[max_idx, 'Typical Price (USD)']

    # Get Cheapest Food for the country
    min_idx = country_data['Typical Price (USD)'].idxmin()
    result['Cheapest Food'] = country_data.loc[min_idx, 'Dish Name']
    result['Cheapest Price (USD)'] = country_data.loc[min_idx, 'Typical Price (USD)']

    results.append(result)

res_df = pd.DataFrame(results)
print(res_df)
```

```python
import pandas as pd

# Display settings
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.width', 0)

# Declaration
# Descriptive Statistic : Total Street Food, Average Price, Most Cooking Method, and Country By Vegetarian Category
data = pd.read_csv("global_street_food.csv")

# Get all unique vegetarian categories
vegetarian = data['Vegetarian'].unique().tolist()

# Prepare container for results
results = []

# Loop through each vegetarian category
for dt in vegetarian:
    # Filter data for this category only
    veg_data = data[data['Vegetarian'] == dt]

    # Count total foods
    total_count = veg_data.shape[0]

    # Calculate average price
    avg_price = veg_data['Typical Price (USD)'].mean()

    # Find most common cooking method
    most_cooking_method = veg_data['Cooking Method'].mode()[0] if not veg_data['Cooking Method'].mode().empty else None

    # Find most common country
    most_country = veg_data['Country'].mode()[0] if not veg_data['Country'].mode().empty else None

    # Store result
    results.append({
        'Vegetarian': dt,
        'Total Food Count': total_count,
        'Average Price (USD)': avg_price,
        'Most Cooking Method': most_cooking_method,
        'Most Country': most_country
    })

# Convert result into DataFrame
res_df = pd.DataFrame(results)

# Sort the data by total food count in descending order
res_df = res_df.sort_values(by='Total Food Count', ascending=False)

print(res_df)
```

```python
import pandas as pd
import statistics
import numpy as np

# Declaration
# Descriptive Statistic : Analyzed Price, Total Vegan Food, and Country With Most Food For Specific Cooking Method
data = pd.read_csv("global_street_food.csv")

country = data['Cooking Method'].unique().tolist()

results = []

for dt in country:
    country_data = data[data['Cooking Method'] == dt]

    total_food = country_data['Typical Price (USD)'].count()
    min_amount = country_data['Typical Price (USD)'].min()
    max_amount = country_data['Typical Price (USD)'].max()
    avg_amount = country_data['Typical Price (USD)'].mean()

    # Count Vegetarian and Non-Vegetarian Food
    total_vegetarian_yes = country_data[country_data['Vegetarian'] == 'Yes'].shape[0]
    total_vegetarian_no = country_data[country_data['Vegetarian'] == 'No'].shape[0]

    # Get Most Found Country in Each Cooking Method
    most_found_country = country_data['Country'].mode()
    most_found_country = most_found_country.iloc[0] if not most_found_country.empty else None
    
    results.append({
        'Cooking Method': dt,
        'Total Food': total_food,
        'Min Typical Price': min_amount,
        'Max Typical Price': max_amount,
        'Average Typical Price': round(avg_amount,2),
        'Total Vegetarian (Yes)': total_vegetarian_yes,
        'Total Non-Vegetarian (No)': total_vegetarian_no,
        'Most Found Country': most_found_country
    })

res_df = pd.DataFrame(results)
res_df = res_df.sort_values(by='Average Typical Price', ascending=False)
res_df
```
